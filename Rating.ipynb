```python
X=df1[['Education',	'JobInvolvement',	'JobLevel',	'MonthlyIncome',	'YearsAtCompany',	'YearsInCurrentRole']]
```

```python
import seaborn as sns
sns.boxplot(X['YearsInCurrentRole'])
```

```python
import numpy as np
Q1=X['MonthlyIncome'].quantile(0.25)
Q3=X['MonthlyIncome'].quantile(0.75)
IQR=Q3-Q1
medians=X['MonthlyIncome'].median()
upper_bound=Q3+1.5*IQR
lower_bound=Q1-1.5*IQR
X['MonthlyIncome']=np.where((X['MonthlyIncome']>upper_bound)|(X['MonthlyIncome']<lower_bound),medians,X['MonthlyIncome'])
```

```python
sns.boxplot(X['MonthlyIncome'])
```

```python
import numpy as np
Q1=X['YearsAtCompany'].quantile(0.25)
Q3=X['YearsAtCompany'].quantile(0.75)
IQR=Q3-Q1
medians=X['YearsAtCompany'].median()
upper_bound=Q3+1.5*IQR
lower_bound=Q1-1.5*IQR
X['YearsAtCompany']=np.where((X['YearsAtCompany']>upper_bound)|(X['YearsAtCompany']<lower_bound),medians,X['YearsAtCompany'])
```

```python
sns.boxplot(X['YearsAtCompany'])
```

```python
import numpy as np
Q1=X['YearsInCurrentRole'].quantile(0.25)
Q3=X['YearsInCurrentRole'].quantile(0.75)
IQR=Q3-Q1
medians=X['YearsInCurrentRole'].median()
upper_bound=Q3+1.5*IQR
lower_bound=Q1-1.5*IQR
X['YearsInCurrentRole']=np.where((X['YearsInCurrentRole']>upper_bound)|(X['YearsInCurrentRole']<lower_bound),medians,X['YearsInCurrentRole'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

le=LogisticRegression()
re=RandomForestClassifier()
```

```python
Y_pred=re.predict(X_test)
```

```python
from sklearn.metrics import classification_report
fulreport=classification_report(Y_test,Y_pred)
```
