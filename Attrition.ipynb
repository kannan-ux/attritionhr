```python
import numpy as np
Q1=X['MonthlyIncome'].quantile(0.25)
Q3=X['MonthlyIncome'].quantile(0.75)
IQR=Q3=Q1
upper_bound=Q3+1.5*IQR
lower_bound=Q1-1.5*IQR
medians=X['MonthlyIncome'].median()
X['MonthlyIncome']=np.where((X['MonthlyIncome']>upper_bound)|(X['MonthlyIncome']<lower_bound),medians,X['MonthlyIncome'])
```

```python
Q1=X['YearsAtCompany'].quantile(0.25)
Q3=X['YearsAtCompany'].quantile(0.75)
IQR=Q3-Q1
upper_bound=Q3+1.5*IQR
lower_bound=Q1-1.5*IQR
medians=X['YearsAtCompany'].median()
X['YearsAtCompany']=np.where((X['YearsAtCompany']>upper_bound)|(X['YearsAtCompany']<lower_bound),medians,X['YearsAtCompany'])
```

```python
Q1=X['YearsAtCompany'].quantile(0.25)
Q3=X['YearsAtCompany'].quantile(0.75)
IQR=Q3-Q1
upper_bound=Q3+1.5*IQR
lower_bound=Q1-1.5*IQR
medians=X['YearsAtCompany'].median()
X['YearsAtCompany']=np.where((X['YearsAtCompany']>upper_bound)|(X['YearsAtCompany']<lower_bound),medians,X['YearsAtCompany'])
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X['MaritalStatus']=le.fit_transform(X['MaritalStatus'])
X['OverTime']=le.fit_transform(X['OverTime'])
```

```python
X=pd.concat([X,pd.get_dummies(X['Department'],dtype="int")],axis=1)
```

```python
X.drop(columns=['Department'],inplace=True)
```

```python

X['MonthlyIncome']=X['MonthlyIncome'].astype("int64")
X['YearsAtCompany']=X['YearsAtCompany'].astype("int64")
```

```python
lt=LabelEncoder()
Y['Attrition']=lt.fit_transform(Y['Attrition'])
```

```python
Y.value_counts()
Y
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
de=DecisionTreeClassifier()
le=LogisticRegression()
de.fit(X_train,Y_train)
```

```python
y_pred=de.predict(X_test)
```

```python
from sklearn.metrics import classification_report
fullreport=classification_report(Y_test,y_pred)
```
