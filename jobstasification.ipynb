```python
import numpy as np
Q1=X['MonthlyIncome'].quantile(0.25)
Q3=X['MonthlyIncome'].quantile(0.75)
IQR=Q3-Q1
upper_bound=Q3+1.5*IQR
lower_bound=Q1-1.5*IQR
medians=X['MonthlyIncome'].median()
X['MonthlyIncome']=np.where((X['MonthlyIncome']>upper_bound)|(X['MonthlyIncome']<lower_bound),medians,X['MonthlyIncome'])
```

```python
Q1=X['TrainingTimesLastYear'].quantile(0.25)
Q3=X['TrainingTimesLastYear'].quantile(0.75)
IQR=Q3-Q1
upperbound=Q3+1.5*IQR
lowerbound=Q1-1.5*IQR
medians=X['TrainingTimesLastYear'].median()
X['TrainingTimesLastYear']=np.where((X['TrainingTimesLastYear']>upperbound)|(X['TrainingTimesLastYear']<lowerbound),medians,X['TrainingTimesLastYear'])

sns.boxplot(X['TrainingTimesLastYear'])
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X['Gender']=le.fit_transform(X['Gender'])
```

```python
X=pd.concat([X,pd.get_dummies(X['Department'],dtype="int")],axis=1)
```

```python
X
```

```python
X.rename(columns={'Human Resources':'Human_Resources','Research & Development':'Research_Development'},inplace=True)	  
```

```python
X=pd.concat([X,pd.get_dummies(X['JobRole'],dtype="int")],axis=1)
```

```python
X
```

```python
X.drop(columns=['Department','JobRole'],inplace=True)
```

```python
X
```

```python
X['MonthlyIncome']=X['MonthlyIncome'].astype("int")
X['TrainingTimesLastYear']=X['TrainingTimesLastYear'].astype("int")
```

```python
X
```

```python
X.rename(columns={'Healthcare Representative':'Healthcare_Representative','Human Resources':'Human_Resources1','Laboratory Technician':'Laboratory_Technician','Manufacturing Director':'Manufacturing_Director','Research Director':'Research_Director','Research Scientist':'Research_Scientist','Sales Executive':'Sales_Executive','Sales Representative':'Sales_Representative'},inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor,DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier
re=RandomForestClassifier()
de=DecisionTreeClassifier()
de.fit(X_train,Y_train)
```

```python
Y_pred=de.predict(X_test)
from sklearn.metrics import classification_report
Accuracy=classification_report(Y_test,Y_pred)
```
